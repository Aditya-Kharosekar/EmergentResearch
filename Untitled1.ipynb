```python
import pandas as pd
import numpy as np
import glob
import json
```

```python
my_claims = pd.read_csv('my_claims_csv_cleaned.csv', index_col = 0)
my_claims.head(5)
```

```python
veracity_dict = {}
claims_file = glob.glob('Snopes\*.json')
index = 1
for l in claims_file:
    item = json.loads(open(l).read())
    veracity_dict[index] = item['Credibility']
    index+=1
```

```python
def label_veracity(row):
    index = row['claimId']
    return veracity_dict[index]
```

```python
my_claims['veracity'] = my_claims.apply (lambda row: label_veracity (row),axis=1)
```
